This script counts valid image files in the **"NORMAL"** folder of the chest X-ray dataset and processes the first image found. It displays the image, along with details about its dimensions, pixel value range, and histogram of pixel intensities.

#### Steps:

1. **Count Image Files**: The function `count_image_files` is used to filter and count image files in a given directory, excluding non-image files like `.DS_Store`.
2. **Load and Display Image**: If there is at least one valid image, it is loaded and displayed using `matplotlib` with a grayscale color map.
3. **Print Image Details**: The script prints out the image's dimensions (height and width in pixels), as well as the maximum and minimum pixel values.
4. **Histogram of Pixel Intensities**: A histogram of the pixel intensities is plotted to show the distribution of pixel values in the image. Additionally, the mean and standard deviation of the pixel intensities are displayed on the plot.

<i><b>Run the next cell to view the image and its pixel distribution:</b></i>


```python
# Function to count files and filter out non-image files
def count_image_files(directory):
    return [
        f
        for f in os.listdir(directory)
        if os.path.isfile(os.path.join(directory, f))
        and f.lower().endswith((".png", ".jpg", ".jpeg", ".bmp", ".tiff"))
    ]


# Get the first image in the "NORMAL" folder, skipping non-image files like .DS_Store
normal_dir = "../input/chest-xray-pneumonia/chest_xray/train/NORMAL"
normal_images = count_image_files(normal_dir)

# Ensure there is at least one image file to process
if normal_images:
    normal_img = normal_images[0]
    sample_img = plt.imread(os.path.join(normal_dir, normal_img))

    # Display the image using matplotlib
    plt.imshow(sample_img, cmap="gray")
    plt.colorbar()
    plt.title("Raw Chest X Ray Image")
    plt.show()

    # Print the image details
    print(f"The dimensions : {sample_img.shape[0]} X {sample_img.shape[1]} pixel.")
    print(
        f"The max pixel value : {sample_img.max():.4f} & the min: {sample_img.min():.4f}"
    )

    # Plot the histogram of pixel intensities
    sns.histplot(
        sample_img.ravel(),
        kde=False,
        color="#0f4c5c",
        label=f"The mean value : {sample_img.mean():.4f} & the std deviation : {sample_img.std():.4f}",
    )

    # Display legend and plot labels
    plt.legend(loc="upper right")
    plt.title("Histogram of Pixel Intensities in the Image")
    plt.xlabel("Pixel Intensity")
    plt.ylabel("No of Pixels")
    plt.show()
else:
    print("No valid images found in the specified directory.")
```

This script sets up an **ImageDataGenerator** for data augmentation, which applies a series of transformations to the images in the dataset. These transformations help improve the model's generalization by artificially increasing the diversity of the data.

### Image Augmentation Parameters:

- **rotation_range**: Randomly rotates the image within the range of 20 degrees.
- **width_shift_range**: Shifts the image horizontally by up to 10% of its width.
- **shear_range**: Applies shear transformations to the image with a maximum of 0.1.
- **zoom_range**: Zooms in or out of the image by up to 10%.
- **samplewise_center**: Centers each image by subtracting the mean pixel value of the image.
- **samplewise_std_normalization**: Normalizes each image by dividing by its standard deviation.

<i><b>Run the next cell to apply these augmentations to the dataset and observe the effects on the images:</b></i>


```python
image_generator = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    samplewise_center=True,
    samplewise_std_normalization=True,
)
```

<p style="padding:10px;background-color:teal;margin:0;color:white;font-family:sans-serif;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Data Preprocessing for Traditional ML Models</p>

Before applying traditional machine learning models like **KNN**, **Logistic Regression**, and **SVM**, we need to preprocess the image data to ensure it's in the correct format. Unlike **CNNs**, which work directly with raw images, traditional machine learning models require images to be flattened into feature vectors, and the pixel values must be normalized.

### Preprocessing Steps:

1. **Load Images from Directories**:

   - Each class of images is stored in a separate subdirectory. We iterate over the directories to load images and their corresponding labels.

2. **Image Resizing**:

   - The images are resized to a fixed size (224x224 in this case) to make them consistent for input into the machine learning models.

3. **Normalization**:

   - Pixel values are normalized by dividing by 255, so they range from 0 to 1. This ensures the models don't get overwhelmed by large pixel values.

4. **Flattening Images**:
   - Images are flattened into a 1D array, as traditional ML models expect each sample to be a feature vector.

### Code Explanation:

The function `load_images_from_directory` performs the following:

- Iterates through the directories containing the image classes.
- Reads each image, resizes it, normalizes the pixel values, and flattens it.
- Appends the flattened image and its corresponding label to the respective lists.

After this step, we have our image data (`X_train`, `X_test`) ready to be used with traditional machine learning classifiers.

<i><b>Run the next cell to load the data and continue with the machine learning models implementation</b></i>


```python
def load_images_from_directory(directory, IMG_SIZE=(224, 224), max_images=None):
    images, labels = [], []
    loaded_count = 0  # Counter for loaded images

    for label, class_name in enumerate(os.listdir(directory)):
        class_dir = os.path.join(directory, class_name)
        if os.path.isdir(class_dir):
            for img_name in os.listdir(class_dir):
                if max_images and loaded_count >= max_images:
                    break
                img_path = os.path.join(class_dir, img_name)
                img = cv2.imread(img_path)
                if img is not None:
                    img = cv2.resize(img, IMG_SIZE)
                    img = img / 255.0  # Normalize the image to [0, 1]
                    images.append(img.flatten())  # Flatten the image into a 1D vector
                    labels.append(label)
                    loaded_count += 1

    return np.array(images), np.array(labels)


# Example usage
train_dir = "../input/chest-xray-pneumonia/chest_xray/train"
test_dir = "../input/chest-xray-pneumonia/chest_xray/test"
val_dir = "../input/chest-xray-pneumonia/chest_xray/val"

# Load images from both NORMAL and PNEUMONIA classes
X_train, y_train = load_images_from_directory(train_dir, max_images=3000)
X_test, y_test = load_images_from_directory(test_dir, max_images=600)
X_val, y_val = load_images_from_directory(val_dir, max_images=18)
print(" Data loaded successfully")
print("Train Data", X_train.shape, y_train.shape)
print("Test Data", X_test.shape, y_test.shape)
print("Validation Data", X_val.shape, y_val.shape)
```

# <p style="padding:10px;background-color:teal;margin:0;color:white;font-family:sans-serif;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Building a CNN Model</p>

In this section, we will explore the process of building a Convolutional Neural Network (CNN) model, which is essential for image classification tasks. We will examine the key concepts involved, such as data preprocessing, model architecture, and the handling of class imbalances to improve model performance.

## Impact of Imbalanced Data on Loss Function

When training a CNN on an imbalanced dataset, the model tends to favor the majority class, leading to poor performance on the minority class. This issue can be addressed by modifying the loss function to take the class imbalance into account. The standard loss function used in binary classification is **cross-entropy** loss:

### Cross-Entropy Loss:

$$\mathcal{L}_{cross-entropy}(x_i) = -(y_i \log(f(x_i)) + (1-y_i) \log(1-f(x_i))),$$

Where:

- \(x_i\) is the input image,
- \(y_i\) is the true label (0 or 1),
- \(f(x_i)\) is the model's predicted probability for the positive class.

The overall average cross-entropy loss over the entire training set \(D\) of size \(N\) is given by:

$$\mathcal{L}_{cross-entropy}(\mathcal{D}) = - \frac{1}{N}\big( \sum_{\text{positive examples}} \log (f(x_i)) + \sum_{\text{negative examples}} \log(1-f(x_i)) \big).$$

However, when we have imbalanced data, this loss function can lead to bias toward the dominant class, as the model may simply predict the majority class to minimize the loss.

### Solution: Weighted Loss Function

To mitigate this bias, we can use a weighted version of the cross-entropy loss function, which assigns higher weights to the minority class and lower weights to the majority class. This balances the contribution to the loss function, helping the model learn from both classes equally.

The weighted cross-entropy loss function is as follows:

$$\mathcal{L}_{cross-entropy}^{w}(x) = - (w_{p} y \log(f(x)) + w_{n}(1-y) \log( 1 - f(x) ) ).$$

Where:

- \(w\_{p}\) is the weight for the positive class (minority),
- \(w\_{n}\) is the weight for the negative class (majority),
- \(f(x)\) and \(y\) are the same as before.

In the following cells, we will implement the CNN model and incorporate this weighted loss function to address the class imbalance problem.


<a id="20-1"></a>

<p style="padding:10px;background-color:teal;margin:0;color:white;font-family:sans-serif;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Class Weight Calculation for Imbalanced Dataset</p>

In this section, we calculate the class weights for the **Normal** and **Pneumonia** classes, addressing the issue of imbalanced data. Imbalanced datasets can cause a CNN model to be biased toward the majority class, which negatively impacts model performance, particularly in detecting the minority class.

- **Class Weights**: We calculate the class weights for **Normal** (class 1) and **Pneumonia** (class 0) using the relative proportions of each class in the **train**, **validation**, and **test** datasets.
- **Visualization**: A pie chart is used to visualize the distribution of these two classes, making it clear how imbalanced the data is.
- **Adjusting Loss Function**: These class weights are essential to modify the loss function, helping the model to focus more on the minority class and achieve balanced performance.

<i><b>This adjustment ensures the CNN model is not biased towards the more frequent class, thus improving its ability to detect pneumonia in chest X-rays.</b></i>


```python
def count_images(directory_list, max_files_per_folder=None):
    """Counts images in directories with an optional limit."""
    total_files = 0

    for directory in directory_list:
        if os.path.exists(directory):
            files = os.listdir(directory)

            # Apply the subset limit if specified
            if max_files_per_folder is not None:
                files = files[:max_files_per_folder]

            total_files += len(files)

    return total_files


# Directories
train_normal = "../input/chest-xray-pneumonia/chest_xray/train/NORMAL"
val_normal = "../input/chest-xray-pneumonia/chest_xray/val/NORMAL"
test_normal = "../input/chest-xray-pneumonia/chest_xray/test/NORMAL"

train_pneumonia = "../input/chest-xray-pneumonia/chest_xray/train/PNEUMONIA"
val_pneumonia = "../input/chest-xray-pneumonia/chest_xray/val/PNEUMONIA"
test_pneumonia = "../input/chest-xray-pneumonia/chest_xray/test/PNEUMONIA"

# Subset size per folder (e.g., 1000 files max per folder)
max_files = 500

# Count files
num_normal = count_images(
    [train_normal, val_normal, test_normal], max_files_per_folder=max_files
)
num_pneumonia = count_images(
    [train_pneumonia, val_pneumonia, test_pneumonia], max_files_per_folder=max_files
)

# Class weights
weight_for_0 = num_pneumonia / (num_normal + num_pneumonia)
weight_for_1 = num_normal / (num_normal + num_pneumonia)
class_weight = {0: weight_for_0, 1: weight_for_1}

# Print weights
print(f"Weight for class 0: {weight_for_0:.2f}")
print(f"Weight for class 1: {weight_for_1:.2f}")

# Pie chart visualization
plt.pie(
    [num_normal, num_pneumonia],
    labels=["Normal", "Pneumonia"],
    colors=["#0f4c5c", "#FFC300"],
    explode=[0.05, 0.05],
    startangle=0,
    autopct="%.1f%%",
)
plt.title("Class Distribution")
plt.show()
```

<a id="20-2"></a>

<p style="padding:10px;background-color:teal;margin:0;color:white;font-family:sans-serif;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Building the CNN Model Architecture</p>

This section describes the architecture of the **Convolutional Neural Network (CNN)** used for classifying chest X-ray images as **Normal** or **Pneumonia**.

### Model Layers:

1. **Input Layer**:
   - The model expects input images of shape **(180, 180, 3)**, where 180x180 represents the image dimensions and 3 channels correspond to the RGB color space.
2. **Convolutional Layers**:

   - **Conv2D layers** are added to detect features such as edges, textures, and patterns in the images.
   - Two sets of **Conv2D layers** are added with filters of sizes **32**, **64**, and **128**, followed by **BatchNormalization** for faster convergence.
   - **MaxPooling** is used after every two convolution layers to reduce the spatial dimensions and retain important features.

3. **Fully Connected Layers**:

   - After the convolutional layers, the model is **flattened** to convert the 2D feature maps into a 1D vector.
   - A **Dense layer** with 128 neurons is added with a **ReLU activation** function to learn non-linear combinations of features.
   - **Dropout** is applied (rate of 0.2) to prevent overfitting during training.

4. **Output Layer**:
   - The final **Dense layer** has one neuron with a **sigmoid activation**, producing a value between 0 and 1. This output represents the probability of the image belonging to the **Pneumonia** class.

### Model Compilation:

- **Loss Function**: Binary Crossentropy is used as the loss function for binary classification.
- **Optimizer**: Adam optimizer is chosen for its adaptive learning rate, which helps speed up training.
- **Metrics**: Accuracy is used as the evaluation metric to monitor the performance of the model.

This model structure ensures robust feature extraction from the chest X-ray images, making it suitable for detecting pneumonia with high accuracy.

<i><b>This CNN architecture is designed to classify chest X-ray images and can be further tuned for better performance using techniques such as data augmentation and learning rate adjustments.</b></i>


```python
# Define the model
model = Sequential()

# First layer: Input layer that defines the input shape
model.add(Input(shape=(180, 180, 3)))

# Adding the Conv2D layers with batch normalization
model.add(Conv2D(filters=32, kernel_size=(3, 3), activation="relu"))
model.add(BatchNormalization())
model.add(Conv2D(filters=32, kernel_size=(3, 3), activation="relu"))
model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2, 2)))

# Adding more Conv2D layers
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation="relu"))
model.add(BatchNormalization())
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation="relu"))
model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2, 2)))

# Adding the final Conv2D layers
model.add(Conv2D(filters=128, kernel_size=(3, 3), activation="relu"))
model.add(BatchNormalization())
model.add(Conv2D(filters=128, kernel_size=(3, 3), activation="relu"))
model.add(BatchNormalization())
model.add(MaxPool2D(pool_size=(2, 2)))

# Flatten the output and add fully connected layers
model.add(Flatten())
model.add(Dense(128, activation="relu"))
model.add(Dropout(0.2))

# Output layer
model.add(Dense(1, activation="sigmoid"))

# Compile the model
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
model.summary()
```

<a id="20-3"></a>

<p style="padding:10px;background-color:teal;margin:0;color:white;font-family:sans-serif;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Training and Validation Data Generators</p>

In this section, we define and configure the data generators used for training and validation of the CNN model. Data generators are essential for handling large datasets that cannot be fully loaded into memory.

### Data Augmentation for Training:

We apply several augmentation techniques to the training data to help the model generalize better by introducing variations into the training images. This helps in improving the robustness of the model. The augmentation includes:

- **Rescaling**: Normalize pixel values to the range [0, 1].
- **Shear Transformation**: Randomly apply shear transformations to the images.
- **Zoom**: Random zoom applied to the images.
- **Rotation**: Random rotations within 15 degrees.
- **Horizontal Flip**: Random horizontal flipping of images.

This augmentation helps prevent overfitting by artificially expanding the training dataset with transformed images.

### Validation Data:

For the validation set, only **rescaling** is applied, as no augmentation is needed. This ensures that the validation set is representative of real-world data without artificial transformations.

### Data Generators:

- **train_datagen** is configured for the training data with all the augmentation techniques.
- **val_datagen** is configured for the validation data, applying only rescaling.

The data generators are created using `flow_from_directory` to load images from directories structured as follows:

- **train_dir**: Directory containing training images in subfolders (Normal, Pneumonia).
- **val_dir**: Directory containing validation images in subfolders (Normal, Pneumonia).

The images are resized to 180x180 pixels to match the input shape of the CNN model.

### Model Training:

The model is trained using the `fit` function, with the following configurations:

- **steps_per_epoch**: The number of batches per epoch, determined by the size of the training dataset.
- **epochs**: The number of times the entire training dataset will be passed through the model.
- **validation_steps**: The number of batches to use for validation.

### Model Evaluation:

After training, the model is evaluated on the validation set to assess its performance. The final loss and accuracy on the validation data are printed.

<i><b>Data augmentation enhances model performance by allowing the model to learn more robust features, while the evaluation provides insights into its generalization ability.</b></i>


```python
# Define the image data generators for training and validation
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,  # Normalize pixel values to the range [0, 1]
    shear_range=0.2,  # Apply random shear transformations
    zoom_range=0.2,  # Apply random zoom transformations
    rotation_range=15,  # Apply random rotations within 15 degrees
    horizontal_flip=True,  # Apply random horizontal flips for augmentation
)

# Validation data generator with only rescaling
val_datagen = ImageDataGenerator(rescale=1.0 / 255)

# Create the data generators
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(180, 180),  # Resize images to match model's expected input size
    batch_size=32,
    class_mode="binary",  # Use binary class mode for binary classification
    shuffle=True,  # Shuffle training data for better generalization
)

validation_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(180, 180),
    batch_size=32,  # Consistent batch size for validation
    class_mode="binary",
    shuffle=False,  # No need to shuffle validation data
)

# Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),  # Number of batches per epoch
    epochs=3,
    validation_data=validation_generator,
    validation_steps=len(validation_generator),
    verbose=1,  # Print progress during training
)

# Evaluate the model on validation data
validation_loss, validation_accuracy = model.evaluate(validation_generator, verbose=1)
print(f"\nValidation Loss: {validation_loss:.4f}")
print(f"Validation Accuracy: {validation_accuracy:.4f}")

# Print the training history
print("\nTraining History:")
for key, values in history.history.items():
    print(f"{key}: {values}")
```
